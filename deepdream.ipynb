## <font color=#0000ff>So what does this have to do with DeepDream?</font>

Just like how you amplified your imagination by over-interpreting a cloud in the sky as a giant drinking a cup of coffee, DeepDream also amplifies a simple image by imaginging different types of features and mixing them into the output image. As we just talked about, this is done by selecting a layer in the CNN (an early one or a later one). Since each layer learns different levels of complexity, we can expect that the surreal type of images generated by DeepDream will vary in terms of what they look like. Some might have entire objects like faces infused onto the image, while other images might simply be amplified by subtle brushstrokes.

#### Let's get (briefly) technical...

(Feel free to skip this section, but it could help you get a sense of what's going on.)

What the DeepDream model is doing behind the scenes, is essentially exciting certain parts of the neural network layers. DeepDream uses a famous type of CNN architecture called ***InceptionV3*** named after the movie Inception. 

If we're interested in say, "Layer 3", what DeepDream will do is get the original image (like a picture of a beautiful gazelle), and make it change in ways that accentuate or amplify all the patterns being discovered in Layer 3. Say Layer 3 is all about brushstroke-type of patterns, with small waves and lines. The dog picture will change so that it will maximize these brushlike patterns. The DeepDream model does this over and over again using a process called Gradient Ascent, that maximizes the chances of the dog picture looking really interesting with brush strokes and wavy lines, versus just being plain and boring. In deep learning, people usually like to go the other way, called Gradient ***Descent***, which means removing all these patterns so that the image the network outputs will be as close as possible to the original dog image - no strange patterns. But with DeepDream, we're all about accentuating and playing up these interesting patterns in the layers we like. What you've just learned are three technical ideas in AI, especially computer vision: 

- ***Activations***: These are the patterns learned and excited at the layer we're interested in. I used the example of brush strokes and wavey lines in Layer 3. It could be any patterns, or features, that any layer has learned and is really excited about. 
- ***Gradients***: These are the ways the original image has changed, so that it contains more of the surrealy dreamy brush strokes and wavey lines. The model is trying to change the way the image of the dog looks, so it plays around with these "gradients" until it gets the image to look surreal. 
- ***Optimization***: If you understand gradient ascent (maximizing the surreal patterns in the image) versus gradient descent (minimizing them), then you get the idea of optimization.

## Step 2 - Process your image

```python
"""
After you have selected an image, by clicking on the cell, 
click and run this cell to download and process it.
"""

# Download an image and read it into a NumPy array.
def download(url, max_dim=None):
    name = url.split('/')[-1]
    image_path = tf.keras.utils.get_file(name, origin=url)
    img = PIL.Image.open(image_path)
    if max_dim:
        img.thumbnail((max_dim, max_dim))
    return np.array(img)

# Normalize an image
def deprocess(img):
    img = 255*(img + 1.0)/2.0
    return tf.cast(img, tf.uint8)

# Display an image
def show(img):
    display.display(PIL.Image.fromarray(np.array(img)))


# Downsizing the image makes it easier to work with.
original_img = download(url, max_dim=500)
show(original_img)
display.display(display.HTML('Image cc-by: <a "href=https://commons.wikimedia.org/wiki/File:Felis_catus-cat_on_snow.jpg">Von.grzanka</a>'))
```

## Step 3 - Prepare the feature extraction model

Download and prepare a pre-trained image classification model. You will use [InceptionV3](https://keras.io/api/applications/inceptionv3/) which is similar to the model originally used in DeepDream. InceptionV3 has been pretrained on a famous image dataset called "ImageNet". You might see this in the code below `weights='imagenet'`. There are many common day objects like cups and cars, as well as animals in the ImageNet dataset. As you continue going through this notebook, see if you see any animals or other shapes in your DeepDream outputs.

```python
base_model = tf.keras.applications.InceptionV3(include_top=False, weights='imagenet')
```

The InceptionV3 architecture is very large. There are 11 layers in InceptionV3, named 'mixed0' though 'mixed10'. Using different layers will result in different dream-like images. Deeper layers respond to higher-level features (such as eyes and faces), while earlier layers respond to simpler features (such as edges, shapes, and textures). Feel free to experiment with the layers selected below. 

For example, if you like, you can change the names in `names = ['mixed3', 'mixed5']`. 

Such as: `names = ['mixed3', 'mixed8']`

***But keep in mind that deeper layers (those with a higher number) will take longer to train on since the gradient computation is deeper.*** 

```python
"""
These are the layers, whose image patterns, we want to "over excite"! The layer names are listed below: mixed3, mixed5.

"""
# Maximize the activations of these layers
names = ['mixed3', 'mixed5']
layers = [base_model.get_layer(name).output for name in names]

# Create the feature extraction model
dream_model = tf.keras.Model(inputs=base_model.input, outputs=layers)
```

## Step 4 - Calculate loss

Remember what we talked about earlier under the "Technical Section" in the introduction? This is where we revisit that idea of how DeepDream really accentuates the image by maximizing the amount of strange patterns on it, versus keeping it just plain and boring. The simple mathematical idea is called "loss".

You may not have learned about what a loss function is, or what the purpose of calculating a loss is all about. For right now, just know that "loss" is the same thing as an "error". ***Usually*** in AI, we want to make the loss, or the error, go down. In other words, we want the neural network to not make any mistakes, thereby decreasing the error of its predictions. However, interestingly enough, in DeepDream we actually ***maximize*** the loss, making it larger. This is a method called "gradient ascent".

```python
"""
Get the image and pass it through our dream_model. This function
will calculate the loss for each layer. In our case we only had two
layers: mixed3 and mixed5.
"""

def calc_loss(img, model):
    img_batch = tf.expand_dims(img, axis=0)
    layer_activations = model(img_batch)
    
    if len(layer_activations) == 1:
        layer_activations = [layer_activations]
    
    losses = []
    for act in layer_activations:
        loss = tf.math.reduce_mean(act)
        losses.append(loss)

    return  tf.reduce_sum(losses)
```

## Step 5 - Gradient ascent

Now that the loss has been calculated, the model must calculate the gradients (weights which are decimal numbers, indicating how much to change the activations), based on the image. DeepDream wants to maximize these surreal looking activations! It's like climbing up a mountain. 

<img src="https://i.pinimg.com/originals/94/59/aa/9459aa149a1b19e2fd69e6115fafd425.gif" height=100 width=100>
Then these gradients are added to the original image. Adding them continually enhances and excites the patterns the network sees. As the image is passed through the network over and over again during training, it will increasingly excite the layers leading to more surreal patterns to emerge. 


```python
class DeepDream(tf.Module):
    
    def __init__(self, model):
        self.model = model

    @tf.function(
        input_signature=(
            tf.TensorSpec(shape=[None,None,3], dtype=tf.float32),
            tf.TensorSpec(shape=[], dtype=tf.int32),
            tf.TensorSpec(shape=[], dtype=tf.float32),)
      )
    
    def __call__(self, img, steps, step_size):
        print("Tracing")
        loss = tf.constant(0.0)
        for n in tf.range(steps):
            with tf.GradientTape() as tape:
                # This needs gradients relative to `img`
                # `GradientTape` only watches `tf.Variable`s by default
                tape.watch(img)
                loss = calc_loss(img, self.model)

            # Calculate the gradient of the loss with respect to the pixels of the input image.
            gradients = tape.gradient(loss, img)

            # Normalize the gradients.
            gradients /= tf.math.reduce_std(gradients) + 1e-8 

            # In gradient ascent, the "loss" is maximized so that the input image increasingly "excites" the layers.
            # You can update the image by directly adding the gradients (because they're the same shape!)
            img = img + gradients*step_size
            img = tf.clip_by_value(img, -1, 1)

        return loss, img
```

```python
"""
We now instantiate the DeepDream model.

"""
deepdream = DeepDream(dream_model)
```

## Step 6 - Main Loop (Hang in there! Almost done!)
Go ahead and run this cell below. It will train the DeepDream model over a series of iterations, calculating the loss (remember, maximizing the surreal patterns) for 50 steps.

```python
"""
This is the main training loop. It calculates the loss for a specific
number of steps, passes through the network. It will take a few
seconds to run through the network, saying "Tracing"

"""
def run_deep_dream_simple(img, steps=50, step_size=0.01): # steps = 100
    # Convert from uint8 to the range expected by the model.
    img = tf.keras.applications.inception_v3.preprocess_input(img)
    img = tf.convert_to_tensor(img)
    step_size = tf.convert_to_tensor(step_size)
    steps_remaining = steps
    
    step = 0
    while steps_remaining:
        if steps_remaining>100:
            run_steps = tf.constant(100)
        else:
            run_steps = tf.constant(steps_remaining)
        steps_remaining -= run_steps
        step += run_steps

        loss, img = deepdream(img, run_steps, tf.constant(step_size))
    
        display.clear_output(wait=True)
        show(deprocess(img))
        print ("Step {}, loss {}".format(step, loss))

    result = deprocess(img)
    display.clear_output(wait=True)
    show(result)

    return result
```

```python
"""
Feel free to modify the "parameters" which are all the arguments in the parentheses.
The run time may increase or decrease depending on what you change. 

You can increase or decrease the steps (iterations for training).
You can increase or decrease the step size.

"""
dream_img = run_deep_dream_simple(img=original_img, steps=50, step_size=0.01) 
```

## Step 7 - Taking it up an octave

Another approach to enhance the image even more is to apply the Gradient Ascent method at different scales. This will allow patterns generated at smaller scales to be incorporated into patterns at higher scales and filled in with additional detail. All you do is apply Gradient Ascent like we just did, and increase the size of the image (which is referred to as an ***octave***), and repeat this process for multiple octaves. ***This will take a minute or so, to run through each iteration and output a new image. Just be patient!***

```python
"""
This may take a minute to run, as it increases the size of the image and runs the image
through the network over again. 
"""
import time
start = time.time()

OCTAVE_SCALE = 1.30

img = tf.constant(np.array(original_img))
base_shape = tf.shape(img)[:-1]
float_base_shape = tf.cast(base_shape, tf.float32)

for n in range(-2, 3):
    # the image sizes up and down using this line of code, ** means to the power of (e.g. 1.3 ^ 3... cubed)
    new_shape = tf.cast(float_base_shape*(OCTAVE_SCALE**n), tf.int32) 
    img = tf.image.resize(img, new_shape).numpy()
    img = run_deep_dream_simple(img=img, steps=50, step_size=0.01)

display.clear_output(wait=True)
img = tf.image.resize(img, base_shape)
img = tf.image.convert_image_dtype(img/255.0, dtype=tf.uint8)
show(img)

end = time.time()
end-start
```

## OPTIONAL - Scaling up with tiles

One thing to consider is that as the image increases in size, so will the time and memory necessary to perform the gradient calculation. The above octave implementation will not work on very large images, or many octaves. To avoid this issue you can split the image into tiles and compute the gradient for each tile. Go ahead and simply run each cell. 

Note: Based on the size of the image and Azure's processing power, running these cells may take a few minutes, be patient!

```python
def random_roll(img, maxroll):
    # Randomly shift the image to avoid tiled boundaries.
    shift = tf.random.uniform(shape=[2], minval=-maxroll, maxval=maxroll, dtype=tf.int32)
    img_rolled = tf.roll(img, shift=shift, axis=[0,1])
    return shift, img_rolled
```

```python
shift, img_rolled = random_roll(np.array(original_img), 512)
show(img_rolled)
```

Here is a tiled equivalent of the `deepdream` function defined earlier:

```python
class TiledGradients(tf.Module):
    def __init__(self, model):
        self.model = model

    @tf.function(
        input_signature=(
        tf.TensorSpec(shape=[None,None,3], dtype=tf.float32),
        tf.TensorSpec(shape=[], dtype=tf.int32),)
      )
    
    def __call__(self, img, tile_size=512):
        shift, img_rolled = random_roll(img, tile_size)
        # Initialize the image gradients to zero.
        gradients = tf.zeros_like(img_rolled)
        
        # Skip the last tile, unless there's only one tile.
        xs = tf.range(0, img_rolled.shape[0], tile_size)[:-1]
        if not tf.cast(len(xs), bool):
            xs = tf.constant([0])
        ys = tf.range(0, img_rolled.shape[1], tile_size)[:-1]
        if not tf.cast(len(ys), bool):
            ys = tf.constant([0])

        for x in xs:
            for y in ys:
                # Calculate the gradients for this tile.
                with tf.GradientTape() as tape:
                    # This needs gradients relative to `img_rolled`.
                    # `GradientTape` only watches `tf.Variable`s by default.
                    tape.watch(img_rolled)

                    # Extract a tile out of the image.
                    img_tile = img_rolled[x:x+tile_size, y:y+tile_size]
                    loss = calc_loss(img_tile, self.model)

                # Update the image gradients for this tile.
                gradients = gradients + tape.gradient(loss, img_rolled)

        # Undo the random shift applied to the image and its gradients.
        gradients = tf.roll(gradients, shift=-shift, axis=[0,1])

        # Normalize the gradients.
        gradients /= tf.math.reduce_std(gradients) + 1e-8 

        return gradients 
```

```python
get_tiled_gradients = TiledGradients(dream_model)
```

Putting this together gives a scalable, octave-aware deepdream implementation:

```python
def run_deep_dream_with_octaves(img, steps_per_octave=50, step_size=0.01, octaves=range(-2,3), octave_scale=1.3):
    base_shape = tf.shape(img)
    img = tf.keras.preprocessing.image.img_to_array(img)
    img = tf.keras.applications.inception_v3.preprocess_input(img)

    initial_shape = img.shape[:-1]
    img = tf.image.resize(img, initial_shape)
    for octave in octaves:
        # Scale the image based on the octave
        new_size = tf.cast(tf.convert_to_tensor(base_shape[:-1]), tf.float32)*(octave_scale**octave)
        img = tf.image.resize(img, tf.cast(new_size, tf.int32))

    for step in range(steps_per_octave):
        gradients = get_tiled_gradients(img)
        img = img + gradients*step_size
        img = tf.clip_by_value(img, -1, 1)

        if step % 10 == 0:
            display.clear_output(wait=True)
            show(deprocess(img))
            print ("Octave {}, Step {}".format(octave, step))
    
    result = deprocess(img)
    return result
```

```python
img = run_deep_dream_with_octaves(img=original_img, step_size=0.01)

display.clear_output(wait=True)
img = tf.image.resize(img, base_shape)
img = tf.image.convert_image_dtype(img/255.0, dtype=tf.uint8)
show(img)
```
